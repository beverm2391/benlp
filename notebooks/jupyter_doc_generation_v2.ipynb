````python
from benlp.llms import Chat, ChatAsync
from benlp.document import Document
import pprint
````

````python
fpath = '/Users/beneverman/Documents/Coding/benlp_v1/data/test/lp_solver.ipynb'
doc = Document(fpath)
doc.process()
data = doc.to_dict()
````

````python
code_cells = [item['text'] for item in data['data'] if item['text'].startswith("'code")]
````

````python
chat = ChatAsync(model='gpt-3.5-turbo')
prompt = """
Write detailed documentation explaining how the code below works.

FORMATTING:

- Use markdown to format your response.
- Use the following markdown to indicate code blocks: ```code```
- Use the following markdown to indicate code blocks with a language: ```python

EXAMPLE:

```python
def add(a, b):
    return a + b
```

This function adds two numbers together.
"""
inputs = [prompt + code_cell for code_cell in code_cells]
outputs = []

await chat.run_chat_async(inputs, outputs)
````

````python
for output in outputs:
    print(output['response'])
````

````python
for idx, output in enumerate(outputs):
    print(f"CODE:\n{code_cells[idx]}\n\nResponse:{output['response']}\n\n")
````

````python
pp = pprint.PrettyPrinter(width=41, compact=True)
for cell in code_cells:
    pp.pprint(cell)
````
